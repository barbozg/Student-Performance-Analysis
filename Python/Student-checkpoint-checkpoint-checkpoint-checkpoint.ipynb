Caracterizando as notas de acordo com o critério a seguir:

 0  - 40 marks : grade E
 
 41 - 60 marks : grade D
 
 60 - 70 marks : grade C
 
 70 - 80 marks : grade B
 
 80 - 90 marks : grade A
 
 90 - 100 marks : grade O

```python
def assign_status(percentage):
    if percentage >= 90:
        return 'O'
    elif percentage >= 80:
        return 'A'
    elif percentage >= 70:
        return 'B'
    elif percentage >= 60:
        return 'C'
    elif percentage >= 40:
        return 'D'
    else:
        return 'E'

df['status'] = df['percentage'].apply(assign_status)

def getgrade(percentage, status):
  if status == 'Fail':
    return 'E'
  if(percentage >= 90):
    return 'O'
  if(percentage >= 80):
    return 'A'
  if(percentage >= 70):
    return 'B'
  if(percentage >= 60):
    return 'C'
  if(percentage >= 40):
    return 'D'
  else :
    return 'E'

df['grades'] = df.apply(lambda x: getgrade(x['percentage'], x['status']), axis = 1 )

df['grades'].value_counts()
```

# Label Encoding

```python
#Encoder é importante para transformar dados categóricos em números

from sklearn.preprocessing import LabelEncoder

# criando encoder
le = LabelEncoder()

# label encoder para teste preparatório
df['test preparation course'] = le.fit_transform(df['test preparation course'])

#Label Encode para o almoço
df['lunch'] = le.fit_transform(df['lunch'])

#label Encode para raca/etnia
df['race/ethnicity'] = df['race/ethnicity'].replace('group A', 1)
df['race/ethnicity'] = df['race/ethnicity'].replace('group B', 2)
df['race/ethnicity'] = df['race/ethnicity'].replace('group C', 3)
df['race/ethnicity'] = df['race/ethnicity'].replace('group D', 4)
df['race/ethnicity'] = df['race/ethnicity'].replace('group E', 5)

#Label Encode para nível de educação
df['parental level of education'] = le.fit_transform(df['parental level of education'])

#Label Encode para gênero
df['gender'] = le.fit_transform(df['gender'])

# label encode para status
df['status'] = le.fit_transform(df['status'])
```

# Data Preparation

```python
x = df.iloc[:,:-1] # todas as colunas, menos a última
y = df.iloc[:,-1] # a última coluna

print(x.shape)
print(y.shape)
```

# Divindo o Data set em training e test

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 45)

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
#MinMaxScaler serve para normalizar dados training

from sklearn.preprocessing import MinMaxScaler

mm = MinMaxScaler()

x_train = mm.fit_transform(x_train)
x_test = mm.transform(x_test)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = None)

x_train = pca.fit_transform(x_train)
x_test = pca.transform(x_test)

explained_variance = pca.explained_variance_ratio_
print(explained_variance)

pca = PCA(n_components = 2)

x_train = pca.fit_transform(x_train)
x_test = pca.transform(x_test)
```

# Modelling

```python
from sklearn.linear_model import LogisticRegression

#Criando o modelo
model = LogisticRegression()

#Alimentando training data para o modelo
model.fit(x_train, y_train)

#prevendo o resultado
y_pred = model.predict(x_test)

#calculando a acurácia
print("Training Accuracy: " , model.score(x_train, y_train))
print("Testing Accuracy: ", model.score(x_test, y_test))
```

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix

#criando uma confusion matrix
cm = confusion_matrix(y_test, y_pred)

#printando ela
plt.rcParams['figure.figsize'] = (8,8)
sns.heatmap(cm, annot = True, cmap = "Greens")
plt.title('Confusion Matrix para a Regressão Logística', fontweight = 30, fontsize = 20)
plt.show()
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

#criando o modelo
model = RandomForestClassifier()

#alimentando o modelo com training data
model.fit(x_train, y_train)

#prevendo o resultado do x_test
y_pred = model.predict(x_test)

#calculando acurácia
print('Training Accuracy: ', model.score(x_train, y_train))
print('Testing Accuracy: ', model.score(x_test, y_test))
```

```python
#criando a confusion matrix
cm = confusion_matrix(y_test, y_pred)

plt.rcParams['figure.figsize'] = (8,8)
sns.heatmap(cm, annot = True, cmap ='Reds')
plt.title('Confusion Matrix para Random Forest')
```
